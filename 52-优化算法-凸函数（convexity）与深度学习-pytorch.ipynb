```python
%matplotlib inline
import numpy as np
import torch
from mpl_toolkits import mplot3d
import dltools
```

```python
f = lambda x: 0.5 * x**2  # 
g = lambda x: torch.cos(np.pi * x)  # 
h = lambda x: torch.exp(0.5 * x)  # 
m = lambda x: torch.maximum(torch.abs(x)-torch.ones_like(x),torch.zeros_like(x)) # 

x, segment = torch.arange(-2, 2, 0.01), torch.tensor([-1.5, 1])
dltools.use_svg_display()
_, axes = dltools.plt.subplots(1, 4, figsize=(9, 3))
for ax, func in zip(axes, [f, g, h,m]):
    dltools.plot([x, segment], [func(x), func(segment)], axes=ax)
```

```python
torch.mean(f(x)) >= f(torch.mean(x))
```

```python
f = lambda x: (x - 1) ** 2
dltools.set_figsize()
dltools.plot([x, segment], [f(x), f(segment)], 'x', 'f(x)')
```
